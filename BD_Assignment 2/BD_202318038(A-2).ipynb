```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
documents = newsgroups.data
len(documents)
```

```python
documents[50]
```

```python
def tfidf_vectorizer(documents):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(documents)
    return vectorizer, tfidf_matrix

def cosine_similarity_between_vectors(vector1, vector2):
    similarity = cosine_similarity(vector1, vector2)
    return similarity[0][0]

def document_similarity_search(query_document, vectorizer, tfidf_matrix, documents, num_results=10):
    query_vector = vectorizer.transform([query_document])
    similarity_scores = cosine_similarity(query_vector, tfidf_matrix)
    ranked_documents = [(score, idx) for idx, score in enumerate(similarity_scores[0])]
    ranked_documents.sort(reverse=True)

    results = []
    for i in range(min(num_results, len(ranked_documents))):
        score, idx = ranked_documents[i]
        results.append((documents[idx], score))

    return results
```

```python
vectorizer, tfidf_matrix = tfidf_vectorizer(documents)
tfidf_matrix.shape
```

```python
query = ("Delving deeper into philosophical debates, questions about the existence of essences, properties of beings,"
         "and the abstract nature of numbers captivate minds. As debates unfold, the intersection of philosophy and technology takes center stage.")
results = document_similarity_search(query, vectorizer, tfidf_matrix, documents, num_results=10)
```

```python
print(f"Query Document: {query}")
print("Similar Documents:")
for doc, score in results:
    print(f"{'*'*50}\nDocument: {doc[:500]}..., \n\nSimilarity Score: {score:.4f}")
```
